# Correlation with Heat Map
This is a very basic technique to derive the most significant input parameter which eventually matters in predicting the output. 


```python
import pandas as pd
import numpy as np
import seaborn as sns

#This is dataset derived from one of the post Si Validation
#Temperature variation(degC) : (-45,-40,-20,0,10,25,50,80,90,105,110)
#voltages(V): (2v97,3v3,3v63)
#Aim is to know which one either Temperature or Voltage is mattering in knowing where significantly failures have occurred. 

data=pd.read_csv('data/post_si_dataset.csv')
data
```

```python
indep_col=data.iloc[:,0:4]
target_col=data.iloc[:,3]
corrmat=data.corr()
top_corr_features = corrmat.index

#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

# Inference
From the above execution we can say that temperature variation(-45deg to 110deg), is highly corelated in the failing scenarios followed by Vdd. 

This methodology can be extended for a huge data collected from Post Si in analyzing the critical input parameters.
